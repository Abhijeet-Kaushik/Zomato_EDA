## Steps For Data analysis

```python
df.isnull().sum()
```

```python
## the temp variables feature will iterate through df.column and checks for the null values and return those columns whose nulll values are grater than 0

[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head(5)
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(10)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## pie chart--top 5 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
## 94.3 %  transaction or records are from india followed by USA and UK
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## observation
1. ratiing b/w 4.5 to 4.9----> Excellent
2. rating b/w 4.0 to 3.9-----> very good
3. rating between 3.5 to3.9----> good
and so on

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python

sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green'])
```

```python
##count plot
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green'])
```
